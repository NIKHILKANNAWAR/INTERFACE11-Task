```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Superstore.csv',encoding='latin-1')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Handling missing values
df.isnull().sum()
```

```python
drop_columns = ['Row ID', 'Customer Name', 'Product Name', 'Postal Code']
df.drop(columns=drop_columns, inplace= True)
```

```python
df.head()
```

```python
if 'Order Date' in df.columns:
    df['Order Date'] = pd.to_datetime(df['Order Date'])
if 'Ship Date' in df.columns:
    df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Category'] = df['Category'].astype('category')
df['Sub-Category'] = df['Sub-Category'].astype('category')
df['Region'] = df['Region'].astype('category')
df['Segment'] = df['Segment'].astype('category')
df['Ship Mode'] = df['Ship Mode'].astype('category')
```

```python
df.info()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='Category', y='Sales', data=df)
plt.title('Sales Distribution by Product Category')
plt.show()
```

```python
df['month'] = df['Order Date'].dt.to_period('M')
sales_trend = df.groupby('month')['Sales'].sum()
plt.figure(figsize=(12, 6))
sales_trend.plot()
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x='Region', y='Sales', hue='Segment', data=df)
plt.title('Sales by Region and Segment')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
num_df = df.select_dtypes(include=np.number)
sns.heatmap(num_df.corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Profit vs. Discount')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

```python
print("Most popular product category:", df.groupby('Category')['Sales'].sum().idxmax())
print("Month with highest sales:", sales_trend.idxmax())
print("Region with highest sales:", df.groupby('Region')['Sales'].sum().idxmax())
print("Segment with highest sales:", df.groupby('Segment')['Sales'].sum().idxmax())
```
