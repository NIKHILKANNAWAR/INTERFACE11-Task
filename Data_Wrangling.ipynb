#Data Wrangling on sales data

##data loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

##Removing unnecessary columns

```python
df.drop(["Row ID", "Order ID", "Customer ID", "Customer Name"], axis = 1, inplace=True)
```

```python
df.head()
```

##Checking Missing values

```python
df.isnull().sum()
```

##Handling missing values

```python
df["Postal Code"].fillna(df.groupby("City")["Postal Code"].transform('first'),inplace=True)
```

```python
df.isnull().sum()
```

##Handling Duplicates

```python
df =df.drop_duplicates()
```

```python
df.shape
```

##Handling inappropriate data types

```python
df.info()
```

##Handling Dates

```python
df["Ship Date"] = pd.to_datetime(df["Ship Date"], format="%d/%m/%Y", errors = 'coerce')
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], format="%d/%m/%Y", errors = 'coerce')
```

```python
df.info()
```

##Handling categorical data

```python
categorical_columns = [col for col in df.columns if df[col].dtypes == 'object']
categorical_columns
```

```python

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in categorical_columns:
  df[col] = le.fit_transform(df[col])
```

```python
df.info()
```

```python
df.head()
```

##Outlier Analysis for Sales

```python
Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - IQR * 1.5
upper_bound = Q3 + IQR * 1.5

df = df[(df["Sales"] > lower_bound) & (df["Sales"] < upper_bound)]
```

```python
df.shape
```

##Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scalar = MinMaxScaler()

df["Sales"]  = scalar.fit_transform(df[["Sales"]])
```

```python
df.head()
```

##Validating sales by checking negative values of sales

```python
df = df[df["Sales"] > 0]
```

```python
df.shape
```
