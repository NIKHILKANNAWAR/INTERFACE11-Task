```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima
from sklearn.metrics import mean_absolute_error, mean_squared_error

df = pd.read_csv("retail_sales.csv")
print(df.head())

df.rename(columns={'data':'date','venda': 'sales', 'estoque': 'stock', 'preco': 'price'}, inplace=True)

df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
df.sort_index(inplace=True)

df = df.resample('M').sum()

df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
df.fillna(method='ffill', inplace=True)
df = df[df['sales'] > 0]

plt.figure(figsize=(12,6))
plt.plot(df.index, df['sales'], label="Sales")
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()

decomp = seasonal_decompose(df['sales'], model='additive', period=12)
decomp.plot()
plt.show()

train_size = int(len(df) * 0.8)
train = df.iloc[:train_size]
test = df.iloc[train_size:]

auto_model = auto_arima(train['sales'], seasonal=True, m=12, trace=True,stepwise=True, suppress_warnings=True)
order = auto_model.order
seasonal_order = auto_model.seasonal_order

sarimax_model = SARIMAX(train['sales'],exog=train[['stock', 'price']],order=order,seasonal_order=seasonal_order,enforce_stationarity=False,enforce_invertibility=False)
model_fit = sarimax_model.fit()

test_pred = model_fit.predict(start=len(train),end=len(train)+len(test)-1,exog=test[['stock', 'price']],dynamic=False)

mae = mean_absolute_error(test['sales'], test_pred)
rmse = np.sqrt(mean_squared_error(test['sales'], test_pred))
print("Test MAE:", mae, " | Test RMSE:", rmse)

plt.figure(figsize=(12,6))
plt.plot(train.index, train['sales'], label="Train Sales", color='blue')
plt.plot(test.index, test['sales'], label="Actual Sales", color='green')
plt.plot(test.index, test_pred, label="Predicted Sales", color='red', linestyle='--')
plt.title("SARIMA: Actual vs Predicted Sales (Test Set)")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()

def forecast_exog_arima(series, steps=6):
    model_exog = ARIMA(series, order=(1,0,0))
    model_exog_fit = model_exog.fit()
    forecast = model_exog_fit.forecast(steps=steps)
    return forecast

future_stock = forecast_exog_arima(df['stock'], steps=6)
future_price = forecast_exog_arima(df['price'], steps=6)
future_exog = pd.DataFrame({'stock': future_stock, 'price': future_price})

future_forecast_obj = model_fit.get_forecast(steps=6, exog=future_exog)
future_pred = future_forecast_obj.predicted_mean

future_dates = pd.date_range(start=df.index[-1] + pd.offsets.MonthEnd(1), periods=6, freq='M')

plt.figure(figsize=(12,6))
plt.plot(df.index, df['sales'], label="Historical Sales", color='blue')
plt.plot(future_dates, future_pred, label="Forecasted Sales", color='red', linestyle='--')
plt.title("Future Sales Forecast")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()

print("Future Sales Predictions:")
print(future_pred)
```
